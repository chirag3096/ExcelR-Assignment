```python
import pandas as pd
import numpy as np
import tweepy
import re 
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
wordnet = WordNetLemmatizer()
import re
from nltk.tokenize import sent_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import requests
from bs4 import BeautifulSoup as bs
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
Iphone_reviews=[]
```

```python
for i in range (1,31):
    op=[]
    url ="https://www.amazon.in/New-Apple-iPhone-Pro-128GB/product-reviews/B08L5V9T31/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"+str(i)
    response = requests.get(url)
    soup = bs(response.content,"html.parser")
    reviews = soup.findAll("span",attrs = {"class","a-size-base review-text review-text-content"})
    for i in range(len(reviews)):
        op.append(reviews[i].text)
    Iphone_reviews = Iphone_reviews+op
```

```python
Iphone_reviews = list(set(Iphone_reviews))
```

```python
txt_upd = ' '.join(Iphone_reviews)
```

```python
txt_upd = re.sub("[^A-Za-z" "]+"," ",txt_upd).lower() 
txt_upd = re.sub("[0-9" "]+"," ",txt_upd).lower() 
txt_upd = re.sub(r'^https?:\/\/.*[\r\n]*', '', txt_upd).lower() 
```

```python
text_tokens = word_tokenize(txt_upd)
```

```python
tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]
```

```python
tf = TfidfVectorizer()
```

```python
text_tf = tf.fit_transform(tokens_without_sw)
```

```python
feature_names = tf.get_feature_names()
dense = text_tf.todense()
denselist = dense.tolist()
df =pd.DataFrame(denselist, columns=feature_names)
```

```python
df
```

```python
word_list = ' '.join(df)
```

```python
wordcloud = WordCloud(background_color='black',
                      width=1800,
                      height=1400).generate(word_list)
```
