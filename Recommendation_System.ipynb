```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data=pd.read_csv('/content/book.csv',encoding='latin-1')
```

```python
data.head()
```

```python
data2=data.iloc[:,1:]
data2
```

```python
data2.sort_values(['User.ID'])
```

```python
len(data2['User.ID'].unique())
```

```python
len(data2['Book.Title'].unique())
```

```python
data3=data2.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
```

```python
data3
```

```python
data3.index=data2['User.ID'].unique()
```

```python
data3
```

```python
data3.fillna(0, inplace=True)
```

```python
data3
```

```python
user_sim= 1 - pairwise_distances(data3.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim1= pd.DataFrame(user_sim)
```

```python
user_sim1
```

```python
user_sim1.index=data2['User.ID'].unique()
user_sim1.columns=data2['User.ID'].unique()
```

```python
np.fill_diagonal(user_sim,0)
user_sim1
```
